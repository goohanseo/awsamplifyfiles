# 새 섹션

```
import os
import sys
import urllib.request
import pandas as pd
import json
# 사용자에게 검색어 입력받기
search_keyword = input("검색어를 입력하세요: \n + : 포함할 단어 \n - : 제외할 단어 \n | : 하나 이상의 단어 검색 \n * : 검색어 모를 때 \n .. : 범위 내 결과 검색 \n "" : 구문 검색 \n 예시) 신한카드 + 하나카드 + 마케팅 \n ")
encText = urllib.parse.quote(search_keyword)

client_id = "reDu1kQzzMJOS0wXHgE9"
client_secret = "Ht9Y_EA3up"

# 검색 결과를 저장할 DataFrame 생성
df = pd.DataFrame(columns=("Title", "Link", "Description", "PubDate"))

# 한 번에 받아올 수 있는 최대 검색 결과 수는 100개이며, start 파라미터를 변경해가며 반복 요청
for start_index in range(1, 1001, 100):  # 1000개 검색 결과를 위한 루프
    url = f"https://openapi.naver.com/v1/search/news?query={encText}&start={start_index}&display=100"  # JSON 결과, 최대 100개

    # API 요청
    request = urllib.request.Request(url)
    request.add_header("X-Naver-Client-Id", client_id)
    request.add_header("X-Naver-Client-Secret", client_secret)

    response = urllib.request.urlopen(request)
    rescode = response.getcode()

    if rescode == 200:
        response_body = response.read()
        response_json = json.loads(response_body.decode('utf-8'))
        items = response_json['items']

        # DataFrame에 검색 결과 추가
        for item in items:
            df = df.append({
                "Title": item['title'],
                "Link": item['link'],
                "Description": item['description'],
                "PubDate": item['pubDate']
            }, ignore_index=True)
    else:
        print("Error Code:" + rescode)
        break

# 결과를 CSV 파일로 저장
csv_filename = "naver_news_search_results.csv"
df.to_csv(csv_filename, index=False, encoding='utf-8-sig')
print(f"{csv_filename} 파일로 저장되었습니다.")
```
